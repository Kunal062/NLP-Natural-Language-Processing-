## Q.1
- Copy the data about the NASA from wikipedia and perform following.
1.  Apply POS Tagging on this data
2.  Remove punctuation symbols
3.  Delete the stopwords
4.  Perform morphological analysis on all the nouns
5.  Find top three words in the data

#### Copy the data about the NASA from wikipedia

```python
# Fetching data from wikipedia

import bs4 as bs
import urllib

source = urllib.request.urlopen('https://en.wikipedia.org/wiki/NASA').read()

soup = bs.BeautifulSoup(source, 'lxml')

text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text

text
```

#### Apply POS Tagging on this data

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
tags = pos_tag(word_tokenize(text))
tags
```

#### Remove punctuation symbols

```python
import string

text_clean = [word for word in word_tokenize(text) if word not in string.punctuation]
text_clean
```

#### Delete Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = stopwords.words('english')
stop_words
```

```python
key_words = [word for word in text_clean if word.lower() not in stop_words]
key_words
```

```python
key_words_tags = pos_tag(key_words)
key_words_tags
```

#### Perform morphological analysis on all the nouns

```python
nouns = [word[0] for word in key_words_tags if word[1].startswith('N')]
nouns
```

#### Find top 3 words in the text

```python
# creating a dictionary to store words from 'text' with their frequencies
count_dict = {}

for i in text_clean:
  if i in count_dict:
    count_dict[i] += 1
  else:
    count_dict[i] = 1
count_dict
```

```python
# creating a sorted (in descending order) version of the dictionary
sorted_dict = sorted(count_dict.items(), key=lambda item: item[1], reverse=True)
sorted_dict
```

```python
# Taking top 3 words of the sorted dictionary
top_3_words = [item[0] for item in sorted_dict[:3]]
top_3_words
```

## Q.2
- Open the Wikipedia page of Pune.
- Find the adjectives used in the text

```python
# Fetching data from the Wikipedia Page
import bs4 as bs
import urllib

source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Pune').read()

soup = bs.BeautifulSoup(source, 'lxml')

text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text

text
```

#### Find the adjectives used in the text

```python
from nltk.tokenize import word_tokenize
import string

text_clean = [word for word in word_tokenize(text) if word not in string.punctuation]
text_clean
```

```python
from nltk import pos_tag
```

```python
text_with_tags = pos_tag(text_clean)
text_with_tags
```

```python
adj = [word[0] for word in text_with_tags if word[1].startswith('J')]
adj
```
