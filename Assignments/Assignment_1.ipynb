### Q.1 
**They told their ages are 23, 25 and 29 respectively.**

- Find the average of ages.
- Count total number of words that starts with 't'/'T'
- Count total punctuation symbols
- Print the words in the capital letter which ends with 'd'
- Merge all the words and print (no punctuations)

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
text = 'They told their ages are 23, 25 and 29 respectively.'
```

#### Find the average of ages.

```python
age = [int(word) for word in word_tokenize(text) if word.isdigit()]
age
```

```python
avg = sum(age)/len(age)
print("The average of ages is : ", avg)
```

#### Count total number of words that starts with 't'/'T'

```python
t_words = [word for word in word_tokenize(text) if word.startswith('t') or word.startswith('T')]
t_words
```

```python
print('The total number of words starting with t/T are : ', len(t_words))
```

#### Count total punctuation symbols

```python
import string
punct = [word for word in text if word in string.punctuation]
punct
```

```python
print('The total number of punctuations in the text are : ', len(punct))
```

#### Print the words in the capital letter which ends with 'd'

```python
d_words = [word.capitalize() for word in word_tokenize(text) if word.endswith('d')]
d_words
```

#### Merge all the words and print (no punctuations)

```python
no_punct = [word for word in word_tokenize(text) if word not in string.punctuation]
''.join(no_punct)
```

### Q.2 
**Open the wikipedia page of Indian Flag and perform following analysis.**

- Count total number of words.
- Calculate percentage of punctuation marks.
- Print total number of lines.
- How many words are in title case.
- Remove the numbers and print the text.

#### Count total number of words.

```python
# Fetch data from Wikipedia

import urllib
import bs4 as bs

# Getting the data source
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Flag_of_India').read()

# Parsing the data/ Creating a beautiful soup object
soup = bs.BeautifulSoup(source, 'lxml')

# Fetch Data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text

text
```

```python
total_tokens = word_tokenize(text)
print(total_tokens)
print("Total number of tokens : ", len(total_tokens))
```

```python
# Removing punctuations
total_words = [word for word in total_tokens if word not in string.punctuation]
total_words
```

```python
print("The total number of words are : ", len(total_words))
```

#### Calculate percentage of punctuation marks.

```python
punct_percent = (len(total_tokens) - len(total_words)) / len(total_tokens)
punct_percent
```

#### Print total number of lines.

```python
from nltk.tokenize import sent_tokenize

total_sent = sent_tokenize(text)
total_sent
```

```python
print("Total number of sentences are : ", len(total_sent))
```

#### How many words are in title case.

```python
titled_words = [word for word in total_words if word.istitle()]
titled_words
```

```python
print("Total number of words in Title case are : ", len(titled_words))
```

#### Remove the numbers and print the text.

```python
no_numbers = [word for word in total_words if not word.isdigit() ]
no_numbers
```
